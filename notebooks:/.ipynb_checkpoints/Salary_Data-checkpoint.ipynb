```python
#Here also there were some issues to open. So, we added extra script on_bad_lines
df = pd.read_csv('salary_data.csv', delimiter=';', on_bad_lines='skip')
#If the CSV uses semicolons (;) instead of commas, and we didn’t specify that, pandas treats the entire row (first row which is the columns headings) as one string.
```

```python
# Convert wage columns to numeric
df['PAID_WAGE_PER_YEAR'] = pd.to_numeric(df['PAID_WAGE_PER_YEAR'], errors='coerce')
df['PREVAILING_WAGE_PER_YEAR'] = pd.to_numeric(df['PREVAILING_WAGE_PER_YEAR'], errors='coerce')

# Convert date columns to datetime
df['CASE_RECEIVED_DATE'] = pd.to_datetime(df['CASE_RECEIVED_DATE'], errors='coerce')
df['DECISION_DATE'] = pd.to_datetime(df['DECISION_DATE'], errors='coerce')

# Convert categorical columns to category type
df['VISA_CLASS'] = df['VISA_CLASS'].astype('category')
df['WORK_STATE'] = df['WORK_STATE'].astype('category')
df['CASE_STATUS'] = df['CASE_STATUS'].astype('category')

# Convert string identifiers to string type
df['CASE_NUMBER'] = df['CASE_NUMBER'].astype(str)
df['EMPLOYER_NAME'] = df['EMPLOYER_NAME'].astype(str)

# Optional: check updated types
print(df.dtypes)
```

```python
# 1. Comparison of JOB TITLE SUBGROUPS
# Highlights the most common job roles.
top_jobs = df['JOB_TITLE_SUBGROUP'].value_counts().nlargest(20)
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title('Top Job Title Subgroups')
plt.xlabel('Number of Cases')
plt.ylabel('JOB TITLE SUBGROUP')
plt.savefig('../images/Top_Job_Title_Subgroups2.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# 2. Average Paid Wage by Visa Class
#Compares wage levels across visa types.

df['PAID_WAGE_PER_YEAR'] = pd.to_numeric(df['PAID_WAGE_PER_YEAR'], errors='coerce')
visa_wage = df.groupby('VISA_CLASS')['PAID_WAGE_PER_YEAR'].mean().sort_values()
visa_wage.plot(kind='barh')
plt.title('Average Paid Wage per Year by Visa Class')
plt.xlabel('Average Wage')
plt.ylabel('Visa Class')
# Save the image plt.save
plt.savefig('Charts/Average_Paid_Wage_per_Year_by_Visa_Class.png', dpi=300, bbox_inches='tight')
# Show the plot 
plt.show() 
```

```python
# 3. Average Paid Wage by Job Title Subgroup
#Compares wage levels across Job Title Subgroup.

df['PAID_WAGE_PER_YEAR'] = pd.to_numeric(df['PAID_WAGE_PER_YEAR'], errors='coerce')
job_title_subgroup_wage = df.groupby('JOB_TITLE_SUBGROUP')['PAID_WAGE_PER_YEAR'].mean().sort_values()
job_title_subgroup_wage.plot(kind='barh')
plt.title('Average Paid Wage Per Year by Job Title Subgroup')
plt.xlabel('Average Wage')
plt.ylabel('Job Title Subgroup')
plt.savefig('Charts/Job_Title_Subgroup.png', dpi=300, bbox_inches='tight')
plt.show() 
```

```python
# 4. Distribution of Prevailing Wage
#Shows wage spread across all cases.

plt.figure(figsize=(20, 10))
df['PREVAILING_WAGE_PER_YEAR'] = pd.to_numeric(df['PREVAILING_WAGE_PER_YEAR'], errors='coerce')
sns.histplot(df['PREVAILING_WAGE_PER_YEAR'].dropna(), bins=50, kde=True)
plt.title('Distribution of Prevailing Wage per Year')
plt.xlabel('Wage')
plt.ylabel('Frequency')
plt.savefig('Charts/Distribution_of_Prevailing_Wage_per_Year.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 5. Case Status Distribution
#Shows how many applications were approved, denied, etc.

sns.countplot(data=df, x='CASE_STATUS')
plt.title('Distribution of Case Status')
plt.xlabel('Case Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('Charts/Distribution_of_Case_Status.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 6. Distribution of Cases by Work State
top_states = df['WORK_STATE'].value_counts().nlargest(10).index
df_top_states = df[df['WORK_STATE'].isin(top_states)]

plt.figure(figsize=(12, 6))
sns.countplot(data=df_top_states, x='WORK_STATE', order=top_states)
plt.title('Top 10 States by Case Count')
plt.xlabel('Work State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Charts/Top_10_States_by_Case_Count.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
# 7. Top Employers with a High Number of Applications
#To focus only on employers with a high number of applications, you can filter your DataFrame based on a threshold 
#— for example, only include employers with at least 500 applications.
employer_counts = df['EMPLOYER_NAME'].value_counts()
#print(employer_counts)
high_volume_employers = employer_counts[employer_counts >= 500].index
#print(high_volume_employers)
df_filtered = df[df['EMPLOYER_NAME'].isin(high_volume_employers)]
#print(df_filtered)
plt.figure(figsize=(16, 8)) 
sns.countplot(data=df_filtered, x='EMPLOYER_NAME', order=high_volume_employers) 
plt.title('Employers with High Number of Applications') 
plt.xlabel('Employer Name') 
plt.ylabel('Application Count') 
plt.xticks(rotation=60) 
plt.savefig('Charts/Employers_with_High_Number_of_Applications.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 8. Top countries by the number of applicants
country_counts = df['COUNTRY_OF_CITIZENSHIP'].value_counts()
#print(country_counts)

high_volume_countries = country_counts[country_counts >=100].index
#print(high_volume_countries)

df_filtered = df[df['COUNTRY_OF_CITIZENSHIP'].isin(high_volume_countries)]

plt.figure(figsize=(16, 8))
sns.countplot(
    data=df_filtered,
    x='COUNTRY_OF_CITIZENSHIP',
    order=high_volume_countries
)

plt.title('Countries with High Number of Applications')
plt.xlabel('Country of Citizenship')
plt.ylabel('Application Count')
plt.xticks(rotation=60)
plt.savefig('Charts/Countries_with_High_Number_of_Applications.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 9. Case Status by Country of Citizenship 
#The following steps analyze case status by country of citizenship. Here we're essentially looking at how visa application outcomes vary across different nationalities.

status_by_country = df.groupby(['COUNTRY_OF_CITIZENSHIP', 'CASE_STATUS']).size().unstack(fill_value=0)
#print(status_by_country.head(10))  # Preview top 10 countries
status_percent = status_by_country.div(status_by_country.sum(axis=1), axis=0)
#print(status_percent.head(10))
top_countries = status_by_country.sum(axis=1).nlargest(10).index
filtered = status_by_country.loc[top_countries]

filtered.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Case Status by Country of Citizenship')
plt.xlabel('Country')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.legend(title='Case Status')
plt.tight_layout()
plt.savefig('Charts/Case_Status_by_Country_of_Citizenship.png', dpi=300, bbox_inches='tight')
plt.show()
```
